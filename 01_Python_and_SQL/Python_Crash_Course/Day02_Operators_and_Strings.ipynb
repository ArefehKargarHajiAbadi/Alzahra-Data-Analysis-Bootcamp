### ۱. عملگرهای ریاضی (Mathematical Operators)
پایتون می‌تواند مانند یک ماشین‌حساب قدرتمند عمل کند. عملگرهای اصلی عبارتند از:
- جمع ($+$) و تفریق ($-$)
- ضرب ($*$) و تقسیم ($/$)
- تقسیم صحیح ($//$): فقط بخش صحیح خارج‌قسمت را برمی‌گرداند.
- باقیمانده ($\%$): باقیمانده تقسیم را برمی‌گرداند.
- توان ($**$)


```python
x = 10
y = 3

print("Addition (+):", x + y)
print("Subtraction (-):", x - y)
print("Multiplication (*):", x * y)
print("Division (/):", x / y)        # نتیجه همیشه float است
print("Floor Division (//):", x // y) # نتیجه بدون اعشار
print("Modulus (%):", x % y)          # باقیمانده تقسیم 10 بر 3
print("Exponentiation (**):", x ** y) # 10 به توان 3
```

### ۲. عملگرهای مقایسه‌ای و منطقی
این عملگرها برای بررسی شروط استفاده می‌شوند و خروجی آن‌ها همیشه $`True` $(درست) یا $`False`$(غلط) است.

**عملگرهای مقایسه‌ای:**
- برابر است با: $==$
- نابرابر است با: $!=$
- بزرگتر/کوچکتر: $>$, $<$
- بزرگتر مساوی/کوچکتر مساوی: $>=$, $<=$

**عملگرهای منطقی:**
- `and`: زمانی $`True` $است که **هر دو** شرط درست باشند.
- `or`: زمانی $`True` $است که **حداقل یکی** از شروط درست باشد.
- `not`: نتیجه شرط را برعکس می‌کند.


```python
a = 15
b = 20

# بررسی عملگرهای مقایسه ای
print("Is a equal to b?", a == b)
print("Is a not equal to b?", a != b)
print("Is a less than b?", a < b)

# بررسی عملگرهای منطقی
condition1 = (a < 20)  # True
condition2 = (b > 30)  # False

print("condition1 AND condition2:", condition1 and condition2) # True and False -> False
print("condition1 OR condition2:", condition1 or condition2)   # True or False -> True
print("NOT condition1:", not condition1)                       # not True -> False
```

### ۳. فرمت‌بندی رشته‌ها (f-strings)
یکی از بهترین ویژگی‌های پایتون (از نسخه 3.6 به بعد)، استفاده از $fstring$ ها برای قرار دادن متغیرها درون متن است. کافیست قبل از شروع رشته حرف $`f`$ را بنویسید و متغیرها را داخل آکولاد `{}` قرار دهید.


```python
first_name = "Arefeh"
age = 25
profession = "Data Analyst"

# استفاده از f-string برای ترکیب متغیرها و متن
profile = f"Hello, my name is {first_name}. I am {age} years old and I work as a {profession}."
print(profile)

# می‌توانید محاسبات را نیز داخل f-string انجام دهید
print(f"In 5 years, {first_name} will be {age + 5} years old.")
```

### ۴. متدهای پرکاربرد رشته‌ها (String Methods)
رشته‌ها در پایتون دارای توابع داخلی $(Method)$ بسیار مفیدی برای پاکسازی و تغییر متن هستند:
- `lower()` و `upper()`: کوچک و بزرگ کردن حروف.
- `replace(old, new)`: جایگزین کردن یک کلمه با کلمه‌ای دیگر.
- `split(separator)`: تبدیل یک رشته به لیستی از کلمات بر اساس یک جداکننده.
- `join(list)`: چسباندن اعضای یک لیست به یکدیگر و ساخت یک رشته.


```python
text = "  Python is awesome for Data Science!  "

# حذف فواصل خالی ابتدا و انتهای رشته (strip) و تبدیل به حروف کوچک
clean_text = text.strip().lower()
print("Cleaned and lowercase:", clean_text)

# جایگزینی کلمات
replaced_text = clean_text.replace("awesome", "perfect")
print("Replaced text:", replaced_text)

# تبدیل رشته به لیست کلمات (بر اساس فاصله)
word_list = replaced_text.split(" ")
print("Split into list:", word_list)

# چسباندن مجدد کلمات با خط تیره
joined_text = "-".join(word_list)
print("Joined with hyphen:", joined_text)
```
