# روز سوم: کنترل جریان برنامه (شرط‌ها)

در این جلسه یاد می‌گیریم که چگونه برنامه‌هایمان را هوشمند کنیم تا بر اساس شرایط مختلف، تصمیمات متفاوتی بگیرند. همچنین با دریافت اطلاعات از کاربر، برنامه‌های خود را تعاملی $(Interactive) $می‌کنیم.

## اهداف امروز:
1. دریافت ورودی از کاربر $(`input`)$
2. دستورات شرطی $(`if`, `elif`, `else`)$
3. شرط‌های تودرتو $(Nested Ifs)$


```python
# دریافت نام کاربر (خروجی رشته است)
user_name = input("Please enter your name: ")
print(f"Hello, {user_name}!")

# دریافت سن کاربر و تبدیل آن به عدد صحیح
age_str = input("Please enter your age: ")
age = int(age_str) # تبدیل رشته به عدد

# اکنون می‌توانیم عملیات ریاضی انجام دهیم
print(f"In 10 years, you will be {age + 10} years old.")
```

### ۲. دستورات شرطی (`if`, `elif`, `else`)
از شرط‌ها برای کنترل جریان برنامه استفاده می‌شود. در پایتون، بلوک‌های کد با فاصله $(Indentation) $مشخص می‌شوند.

- `if`: شرط اصلی را بررسی می‌کند. اگر درست باشد $(True)$، کدهای داخل آن اجرا می‌شوند.
- `elif`: مخفف $else if$ است. اگر شرط قبلی غلط بود، این شرط بررسی می‌شود.
- `else`: اگر هیچ‌کدام از شروط بالا درست نبود، این بخش اجرا می‌شود.


```python
score = 85

# بررسی نمره و تخصیص گرید
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: F")
```

### ۳. شرط‌های تودرتو (Nested Ifs)
گاهی اوقات نیاز داریم داخل یک شرط، شرط دیگری را نیز بررسی کنیم. به این ساختار، شرط تودرتو می‌گویند.
در این حالت باید به میزان تورفتگی $(Indentation)$ کدها دقت کنید.


```python
number = 15

# ابتدا بررسی می‌کنیم آیا عدد مثبت است؟ $number > 0$
if number > 0:
    print("The number is positive.")
    
    # حالا بررسی می‌کنیم آیا زوج است یا فرد؟ $number \% 2 == 0$
    if number % 2 == 0:
        print("It is an EVEN number.")
    else:
        print("It is an ODD number.")
        
elif number == 0:
    print("The number is ZERO.")
else:
    print("The number is negative.")
```
