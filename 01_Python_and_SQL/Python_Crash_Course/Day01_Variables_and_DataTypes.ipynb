## ۲. متغیرها (Variables)
متغیرها مانند جعبه‌هایی هستند که داده‌ها را در خود ذخیره می‌کنند. در پایتون نیازی به تعریف نوع متغیر پیش از مقداردهی نیست.
به عنوان مثال در رابطه ریاضی $ x = 10 $، مقدار ده درون متغیر ایکس قرار می‌گیرد.

**قوانین نام‌گذاری استاندارد:**
- نام متغیر فقط می‌تواند شامل حروف، اعداد و خط تیره پایین (Underscore) باشد.
- نام متغیر نمی‌تواند با عدد شروع شود (مثلاً `1message` غلط است).
- در پایتون معمولاً از حروف کوچک و خط تیره پایین برای کلمات چندبخشی استفاده می‌شود (مثل `student_name`).


```python
# تعریف متغیرها
message = "Welcome to the bootcamp!"
print(message)

# تغییر مقدار متغیر (پایتون همیشه آخرین مقدار را در نظر می‌گیرد)
message = "Let's learn Python."
print(message)

# ترکیب متغیرها در تابع پرینت
first_name = "Ali"
last_name = "Rezaei"
print(first_name, last_name)
```

## ۳. انواع داده‌های پایه (Data Types)
در این مرحله با ۴ نوع داده اصلی که در علم داده بسیار کاربرد دارند آشنا می‌شویم:
1. **Integer (int):** اعداد صحیح (مانند $15$ یا $-3$)
2. **Float (float):** اعداد اعشاری (مانند $3.14$ یا $2.5$)
3. **String (str):** رشته‌های متنی که داخل کوتیشن `' '` یا دابل کوتیشن `" "` قرار می‌گیرند.
4. **Boolean (bool):** مقادیر منطقی که فقط می‌توانند درست $(`True`)$ یا غلط $(`False`) $باشند.

برای تشخیص نوع یک متغیر، از تابع $`type()`$ استفاده می‌کنیم.


```python
# 1. Integer (اعداد صحیح)
age = 25
print("Age:", age)
print("Type of Age:", type(age))

print("-" * 20) # چاپ خط جداکننده

# 2. Float (اعداد اعشاری)
temperature = 36.5
print("Temperature:", temperature)
print("Type of Temperature:", type(temperature))

print("-" * 20)

# 3. String (رشته متنی)
course_name = "Data Science"
print("Course:", course_name)
print("Type of Course:", type(course_name))

print("-" * 20)

# 4. Boolean (مقدار منطقی)
# دقت کنید که T و F باید با حروف بزرگ نوشته شوند
is_student = True 
print("Is Student?", is_student)
print("Type of is_student:", type(is_student))
```
