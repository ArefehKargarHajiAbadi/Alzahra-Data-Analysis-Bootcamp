# روز چهارم: تکرار و حلقه‌ها $(Loops)$

در برنامه‌نویسی، اغلب نیاز داریم یک قطعه کد را چندین بار اجرا کنیم. حلقه‌ها $(Loops)$ به ما این امکان را می‌دهند که از تکرار کدهای مشابه جلوگیری کنیم و کدی تمیزتر و کارآمدتر بنویسیم. در پایتون دو نوع حلقه اصلی داریم: $`for` $و $`while`$.

## اهداف امروز:
1. آشنایی با حلقه $`for`$ و تابع $`range`$
2. آشنایی با حلقه $`while`$ و جلوگیری از حلقه‌های بی‌نهایت
3. $`break`$ $`continue`$ کنترل جریان حلقه با استفاده از 


### ۱. حلقه $`for`$ و تابع $`range()`$
حلقه $`for`$ برای تکرار روی یک توالی (مانند لیست‌ها، رشته‌ها یا محدوده‌ای از اعداد) استفاده می‌شود. 
تابع $`range()`$ یک توالی از اعداد تولید می‌کند و ساختار کلی آن به شکل $`range(start, stop, step)`$ است.
دقت کنید که عدد $stop$ در بازه قرار نمی‌گیردتا $n-1$  پیش می‌رود.


```python
# چاپ اعداد از $0$ تا $4$ (عدد $5$ شامل نمی‌شود)
print("--- range(5) ---")
for i in range(5):
    print(i)

# چاپ اعداد از $2$ تا $8$ با گام $2$
print("\n--- range(2, 9, 2) ---")
for j in range(2, 9, 2):
    print(j)

# تکرار روی کاراکترهای یک رشته
print("\n--- String Iteration ---")
word = "Data"
for char in word:
    print(char)
```

### ۲. حلقه `while`
حلقه $`while`$ تا زمانی که یک شرط خاص درست $(`True`)$ باشد، به اجرای کد ادامه می‌دهد. 

**هشدار (حلقه بی‌نهایت):** اگر شرط پایان حلقه هرگز نادرست$`False`$نشود، برنامه تا ابد به کار خود ادامه می‌دهد ($Infinite Loop$). همیشه مطمئن شوید که متغیرهای کنترل‌کننده شرط در داخل حلقه به‌روزرسانی می‌شوند.


```python
# یک شمارنده ساده با حلقه while
counter = 1

while counter <= 3:
    print(f"Counter is at: {counter}")
    # به‌روزرسانی شمارنده برای جلوگیری از حلقه بی‌نهایت
    counter += 1 

print("Loop finished!")
```

### ۳. کلمات کلیدی $`break`$ $`continue`$
گاهی نیاز داریم رفتار پیش‌فرض حلقه را تغییر دهیم:
- $`break`$: به طور کامل حلقه را متوقف کرده و از آن خارج می‌شود.
- $`continue`$: از اجرای کدهای باقی‌مانده در دور فعلی صرف‌نظر کرده و مستقیماً به دور بعدی ($Iteration $بعدی) می‌رود.


```python
print("--- Using break ---")
for num in range(1, 10):
    if num == 5:
        print("Breaking the loop at $5$!")
        break
    print(num)

print("\n--- Using continue ---")
for num in range(1, 6):
    if num == 3:
        print("Skipping number $3$!")
        continue # رفتن به تکرار بعدی
    print(num)
```
